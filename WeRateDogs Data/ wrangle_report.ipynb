### About the Project
WeRateDogs is a popular Twitter account that rates people's dogs with a humorous comment about the dog. These ratings almost always have a denominator of 10. The numerators, though? Almost always greater than 10. 11/10, 12/10, 13/10, etc. Why? Because "they're good dogs Brent." WeRateDogs has over 4 million followers and has received international media coverage.

In this project we have three sources of data to consume from and are required to gather, assess, clean, Store, analyzing, and finally visualizing wrangled data.


### Data Wrangling
#### Gathering

Data was gathered from three sources: 
1.	twitter_archive_enhanced.csv provided by Udacity
2.	image_predictions.tsv hosted on udacity servers and was to be downloaded programmatically
3.	Tweet.json file which was to be obtained through a twitter API using the Tweepy library

#### Assessing
I programmatically assessed all three databases to quality and tidiness issues in which I then documented the below before cleaning.

#### Quality

#### Twitter archive
•	Timestamp column should be in time format and not as object.
•	We only need user ratings and not retweets and the data set has some unnecessary columns.
•	Source column data is unreadable HTML.

#### Twitter Images
•	Images, p1_conf, p2_conf and p3_conf aren’t headers properly names.
•	p1, p2, and p3 values aren’t consistently named.
•	Values for p1, p2, and p3 are breed types so it would be ideal if they were in category format.
•	Tweet ID’s should be strings and not Integers.

#### Twitter Json
•	Delete unnecessary columns.

#### Tidiness
•	The doggo, floofer, pupper and puppo columns should be one column and as categories.
•	Merge all cleaned datasets for visualization

#### Cleaning and storing
In this section I started of by creating copies of all three datasets which was a very important step as I continued to work on the dataset, whenever I made a mistake I could always recall from the original dataset (Thank you Udacity Instructors!).

From the above issues I followed three steps to solve them by using code.
1.	Define: This consisted definition for the solution
2.	Code: Coding the solution
3.	Test: This phase was to test if the coding function worked.

During this stage I had major issues fixing the data, which led me to more research as well as ask peers, which led me to read on more documentation on pandas, seaborn (although this was mostly during visualization) as well as figure out how to use the pandas cheatcodes.

#### Storing Data
Finally after clearing quality and tidiness issues, I merged all the three datasets and stored then in a master dataset named twitter_archive_master.csv.
